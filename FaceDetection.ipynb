```python
import os
import time
import uuid
import cv2
```

### image collection

```python
imgPath = os.path.join('data', 'images')
imgNum = 13
time.sleep(2)
cam = cv2.VideoCapture(0)
for i in range(imgNum):
    print(f"Collecting image {i}")
    ret, frame = cam.read() # ret is 0 or 1 showing whether video capture was successfull
    imgName = os.path.join(imgPath, f'{str(uuid.uuid1())}.jpg')
    cv2.imwrite(imgName, frame) #save img
    cv2.imshow('Frame: ', frame) 
    time.sleep(0.75)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cam.release()
cv2.destroyAllWindows()
```
